```python
import re
import warnings; warnings.filterwarnings("ignore")

def calculate_mp_sum(camera_string):

    camera_string = camera_string.replace('Million Effective Pixel', 'MP')
    
    mp_values = re.findall(r'(\d+(\.\d+)?)\s*MP', camera_string)
    
    if mp_values:
        mp_sum = sum(int(float(value[0])) for value in mp_values)
    else:
        mp_sum = 0
    return mp_sum


df['Back_Camera'] = df['Back_Camera'].apply(calculate_mp_sum) 
```

```python

def calculate_mp_sum(camera_string):
    if camera_string:  
        
        mp_values = re.findall(r'(\d+(\.\d+)?)\s*MP', camera_string)
        
        if mp_values:
            mp_sum = sum(int(float(value[0])) for value in mp_values)
        else:
            mp_sum = 0
    else:
        mp_sum = 0
    return mp_sum


df['Front_Camera'] = df['Front_Camera'].apply(calculate_mp_sum) 
```

```python

def split_display(display_string):
    
    display_size = re.search(r'(\d+(\.\d+)?)\s*cm', display_string)
    
    display_type = re.search(r'(?<=inch\)).*', display_string)
    
    size = display_size.group(1) if display_size else None
    type_ = display_type.group().strip() if display_type else None
    
    return size, type_


df['Display_Size'], df['Display_Type'] = zip(*df['display'].apply(split_display))
```

```python

def standardize_display_type(display_type):
    if display_type is None or 'Display' == display_type:
        return 'Unknown'
    elif 'HD+' in display_type:
        return 'HD'
    elif 'HD' in display_type and 'Full' in display_type:
        return 'Full HD'
    elif 'Quad HD' in display_type:
        return 'Quad HD'
    elif 'Super AMOLED' in display_type:
        return 'AMOLED'
    elif 'quarter HD' in display_type:
        return 'Quarter HD'
    elif 'FWVGA' in display_type:
        return 'FWVGA'
    elif 'WVGA' in display_type:
        return 'WVGA'
    elif 'HVGA' in display_type:
        return 'HVGA'
    else:
        return 'Generic Display'


df['Display_Type'] = df['Display_Type'].apply(standardize_display_type) 
```

```python

def extract_memory_specs(memory_string):
    parts = memory_string.split('|')  
    
    ram = parts[0].strip() if len(parts) > 0 else '0 GB RAM' 
    rom = parts[1].strip() if len(parts) > 1 else '0 GB ROM' 
    expandable = parts[2].replace('Expandable Upto', '').strip() if len(parts) > 2 else '0 GB'
    
    return ram, rom, expandable 
```

```python
df[['RAM', 'ROM', 'Expandable_Memory']] = df['memory'].apply(lambda x: pd.Series(extract_memory_specs(x))) 
```

```python
def convert_ram_to_gb(ram_string):
    
    numeric_value, unit = ram_string.split()[0], ram_string.split()[1]

    
    numeric_value = float(numeric_value)

    
    if unit == "MB":
        numeric_value /= 1024


    return int(numeric_value)


df['RAM'] = df['RAM'].apply(convert_ram_to_gb) 
```

```python

def convert_to_gb(memory_string):
    if pd.isna(memory_string): 
        return 0


    parts = memory_string.split()
    numeric_value = float(parts[0])
    
    
    if len(parts) > 1:
        unit = parts[1]
        if unit == "TB":
            numeric_value *= 1024  
        elif unit == "MB":
            numeric_value /= 1024  


    return int(numeric_value)


df['Expandable_Memory'] = df['Expandable_Memory'].apply(convert_to_gb)
```

```python
def extract_processor_name(text):
    
    if isinstance(text, str):
        
        pattern = r'([A-Za-z0-9]+(?: [A-Za-z0-9]+)*(?: [A-Za-z0-9]+)* Processor(?: [0-9.]+ GHz)?(?: \(.*\))?)'
        match = re.search(pattern, text)
        return match.group(0) if match else None
    else:
        return None


df['processor'] = df['processor'].apply(extract_processor_name) 
```

```python
df['processor'].unique()  
```

```python

def categorize_processor(processor_info):
    if isinstance(processor_info, str):  
        processor_info = processor_info.lower() 
        if 'qualcomm' in processor_info or 'snapdragon' in processor_info:
            return 'Qualcomm'
        elif 'mediatek' in processor_info or 'helio' in processor_info or 'mtk' in processor_info:
            return 'MediaTek'
        elif 'samsung' in processor_info or 'exynos' in processor_info:
            return 'Exynos'
        elif 'hisilicon' in processor_info or 'kirin' in processor_info:
            return 'HiSilicon'
        elif 'spreadtrum' in processor_info or 'cortex' in processor_info:
            return 'Spreadtrum'
        else:
            return 'Unknown'
    else:
        return 'Unknown' 

df['processor'] = df['processor'].apply(categorize_processor) 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


def plot_univariate(df, column_name):
    plt.figure(figsize=(10, 4))
    if df[column_name].dtype == 'object':
        sns.countplot(data=df, y=column_name, order=df[column_name].value_counts().index)
        plt.title(f'Count Plot for {column_name}')
        plt.xlabel('Count')
    else:
        sns.histplot(df[column_name], kde=True, bins=30)
        plt.title(f'Distribution Plot for {column_name}')
        plt.xlabel(column_name)
        plt.ylabel('Frequency')  
    plt.show() 
```

```python

for col in numeric_columns:
    print(f"\nStatistics for {col}:\n{df[col].describe()}")
    plot_univariate(df, col) 
```
